### Predictive Modelling

```python
#LOGISTIC REGRESSION
# lets split the Dataset for Predictive Modelling

y = data['label']
x = data.drop(['label'], axis = 1)

print("Shape of x:", x.shape)
print("Shape of y:", y.shape)
```

```python
# lets create Training and Testing Sets for Validation of Results
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state = 0)
print(x_train)

print("The Shape of x train:", x_train.shape)
print("The Shape of x test:", x_test.shape)
print("The Shape of y train:", y_train.shape)
print("The Shape of y test:", y_test.shape)
```

```python
# lets create a Predictive Model

from sklearn.linear_model import LogisticRegression
#model = LogisticRegression(solver='lbfgs')
model = LogisticRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
# lets evaluate the Model Performance
from sklearn.metrics import classification_report

# lets print the Classification Report also
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
from matplotlib import pyplot as plot
plot.scatter(y_test, y_pred, label='test', c="yellow",edgecolor ="black")
plot.title('Logistic Regression')
plot.xlabel('Actual crop')
plot.xticks( rotation='vertical')
plot.ylabel('Predicted crop')
plot.show()
```

```python
#calculating diff types of errors
from sklearn import metrics
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
#for column_name in test_data.columns:
D3=y_test
F3=y_pred
if D3.dtype == object:
    D3 = le.fit_transform(D3)
else:
    pass
if F3.dtype == object:
    F3 = le.fit_transform(F3)
else:
    pass
print('Mean Absolute Error:', metrics.mean_absolute_error(D3, F3))
print('Mean Squared Error:', metrics.mean_squared_error(D3, F3))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(D3, F3)))
```

```python
from sklearn.metrics import accuracy_score
accu=accuracy_score(D3, F3)
print('Accuracy for Logistic Regression is ',accu)
```

```python
from sklearn.model_selection import ParameterGrid
import numpy as np

# Create a dictionary of hyperparameters to search
# n_estimators is the number of trees in the forest. The larger the better, 
# but also the longer it will take to compute. 
# Run grid search

grid = {'n_estimators': [1000], 'max_depth': [500], 'max_features': [5], 'random_state': [7]}
test_scores = []

# Loop through the parameter grid, set the hyperparameters, and save the scores
for g in ParameterGrid(grid):
    rfr.set_params(**g)  # ** is "unpacking" the dictionary
    rfr.fit(train_feature2, train_target2)
    test_scores.append(rfr.score(test_feature2, test_target2))

# Find best hyperparameters from the test score and print
best_idx = np.argmax(test_scores)
print(test_scores[best_idx], ParameterGrid(grid)[best_idx])  # You don't want negative value

# The best test score
```

```python
# Use the best hyperparameters from before to fit a random forest model
rfr = RandomForestRegressor(n_estimators=1000, max_depth=500, max_features = 5, random_state=7)
rfr.fit(train_feature2, train_target2)

# Make predictions with our model
train_prediction2 = rfr.predict(train_feature2)
test_prediction2 = rfr.predict(test_feature2)
```

```python
# Find Best max depth

# Loop through a few different max depths and check the performance
# Try different max depths. We want to optimize our ML models to make the best predictions possible.
# For regular decision trees, max_depth, which is a hyperparameter, limits the number of splits in a tree.
# You can find the best value of max_depth based on the R-squared score of the model on the test set.

for d in [2,5,7,8,10,15,50,100,500,1000]:
    # Create the tree and fit it
    dr= DecisionTreeRegressor(max_depth=d)
    dr.fit(train_feature2, train_target2)

    # Print out the scores on train and test
    print('max_depth=', str(d))
    print(dr.score(train_feature2, train_target2))
    print(dr.score(test_feature2, test_target2), '\n')  
    # You want the test score to be positive and high
```
